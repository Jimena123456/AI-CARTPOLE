### Definir las funciones

```python
from collections import Counter
import numpy as np
from random import choices

# Matriz de adyacencia
adyacencia = np.zeros((len(estaciones), len(estaciones)), dtype=int)

# Definición de las conexiones
conexiones = conexiones

# Llenado de la matriz de adyacencia
for conexion in conexiones:
    origen = estaciones[conexion[0]]
    destino = estaciones[conexion[1]]
    adyacencia[origen][destino] = 1
    adyacencia[destino][origen] = 1  # Consideramos la conexión bidireccional

# Considerar las distancias
distancias = distancias

# Asignar distancias a la matriz de adyacencia
matriz_dist = np.zeros((len(estaciones), len(estaciones)), dtype=int)

for origen, destino, distancia in distancias:
    indice_origen = estaciones[origen]
    indice_destino = estaciones[destino]
    matriz_dist[indice_origen][indice_destino] = distancia
    matriz_dist[indice_destino][indice_origen] = distancia  # Considera la conexión bidireccional

# Hacer la función fitness
def fitness(genome):
    total_distance = 0  # Inicializar la variable para almacenar la distancia total
    for i in range(len(genome) - 1):
        total_distance += matriz_dist[genome[i]][genome[i + 1]]  # Sumar la distancia entre cada estación consecutiva
    return total_distance  # Devolver la distancia total calculada

# Hacer un genoma
def create_genome(gen_size):
    genome = [estaciones["El Rosario"]]  # Empezar con la estación "El Rosario"
    while True:
        current_station = genome[-1]  # Obtener la última estación en el genoma
        next_station_candidates = np.where(adyacencia[current_station] == 1)[0]  # Obtener las estaciones conectadas a la actual
        next_station = np.random.choice(next_station_candidates)  # Elegir aleatoriamente una estación conectada como siguiente
        genome.append(next_station)  # Agregar la siguiente estación al genoma
        if next_station == estaciones["San Lázaro"]:  # Verificar si se llegó a la estación "San Lázaro"
            break  # Si se llegó a "San Lázaro", salir del bucle
    return genome  # Devolver el genoma generado

# Definir algunos parámetros
city_no = len(adyacencia) - 1
pop_size = 10
gen_size = 40
epochs = 600
elitism_coef = 2

# Crear un genoma inicial
genome = create_genome(gen_size)
solution = [names[i] for i in genome]  # Convertir el genoma en una lista de nombres de estaciones
```
